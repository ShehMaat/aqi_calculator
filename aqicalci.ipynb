```python
#ce_sulfur,ce_PM10,ce_PM2,ce_NO,ce_NO2,ce_CO=0,0,0,0,0,0
volume_cement=0
#calculation for cement 
def Cement():
    global ce_sulfur
    global ce_PM10
    global ce_PM2
    global ce_NO
    global ce_NO2
    global ce_CO
    global volume_cement
    cement=eval(input("ENTER THE QUANTITY OF CEMENT USED IN kg"))
    volume_cement=cement/1440000
    ce_sulfur=volume_cement*0.02*1000000
#weight of sulfur calculated for cement
    ce_PM10=volume_cement*0.3*1000000
#weight of PM10 calculated for cement
    ce_PM2=volume_cement*0.15*1000000
#weight of PM2.5 calculated for cement
    ce_NO=volume_cement*0.1*1000000
#weight of NO calculated for cement
    ce_NO2=volume_cement*0.05*1000000
#weight of NO2 calculated for cement
    ce_CO=volume_cement*0.1*1000
#weight of CO calculated for cement
```

```python
#calculation for steel 
def Steel():
    global s_sulfur
    global s_PM10
    global s_PM2
    global s_NO
    global s_NO2
    global s_CO
    global volume_steel
    steel=eval(input("ENTER THE QUANTITY OF STEEL USED IN kg"))
    volume_steel=steel/7850000
    s_sulfur=volume_steel*0.05*1000000
#weight of sulfur calculated for steel
    s_PM10=volume_steel*0.8*1000000
#weight of PM10 calculated for steel
    s_PM2=volume_steel*0.4*1000000
#weight of PM2.5 calculated for steel
    s_NO=volume_steel*0.2*1000000
#weight of NO calculated for steel
    s_NO2=volume_steel*0.1*1000000
#weight of NO2 calculated for steel
    s_CO=volume_steel*0.15*1000
#weight of CO calculated for steel
```

```python
#calculation for asphalt
def Asphalt():
    global a_sulfur
    global a_PM10
    global a_PM2
    global a_NO
    global a_NO2
    global a_CO
    global volume_asphalt
    asphalt=eval(input("ENTER THE QUANTITY OF ASPHALT USED IN kg"))
    volume_asphalt=asphalt/2350000
    a_sulfur=volume_asphalt*0.05*1000000
#weight of sulfur calculated for asphalt
    a_PM10=volume_asphalt*0.5*1000000
#weight of PM10 calculated for asphalt
    a_PM2=volume_asphalt*0.25*1000000
#weight of PM2.5 calculated for asphalt
    a_NO=volume_asphalt*0.1*1000000
#weight of NO calculated for asphalt
    a_NO2=volume_asphalt*0.05*1000000
#weight of NO2 calculated for asphalt
    a_CO=volume_asphalt*0.2*1000
#weight of CO calculated for asphalt
```

```python
#calculation for diesel
def Diesel():
    global d_sulfur
    global d_PM10
    global d_PM2
    global d_NO
    global d_NO2
    global d_CO
    global volume_diesel
    diesel=eval(input("ENTER THE QUANTITY OF DIESEL USED IN kg"))
    volume_diesel=diesel/850000
    d_sulfur=volume_diesel*0.03*1000000
#weight of sulfur calculated for diesel
    d_PM10=volume_diesel*0.8*1000000
#weight of PM10 calculated for diesel
    d_PM2=volume_diesel*0.4*1000000
#weight of PM2.5 calculated for diesel
    d_NO=volume_diesel*0.3*1000000
#weight of NO calculated for diesel
    d_NO2=volume_diesel*0.15*1000000
#weight of NO2 calculated for diesel
    d_CO=volume_diesel*0.3*1000
#weight of CO calculated for diesel
```

```python
#calculation for lpg
def LPG():
    global l_PM10
    global l_PM2
    global l_NO
    global l_NO2
    global l_CO
    global volume_lpg
    lpg=eval(input("ENTER THE QUANTITY OF LPG USED IN kg"))
    volume_lpg=lpg/510000
    l_PM10=volume_lpg*0.02*1000000
#weight of PM10 calculated for lpg
    l_PM2=volume_lpg*0.4*1000000
#weight of PM2.5 calculated for lpg
    l_NO=volume_lpg*0.08*1000000
#weight of NO calculated for lpg
    l_NO2=volume_lpg*0.04*1000000
#weight of NO2 calculated for lpg
    l_CO=volume_lpg*0.08*1000
#weight of CO calculated for lpg
```

```python
#calculation for gasoline
def Gasoline():
    global g_PM10
    global g_PM2
    global g_NO
    global g_NO2
    global g_CO
    global volume_gasoline
    gasoline=eval(input("ENTER THE QUANTITY OF GASOLINE USED IN kg"))
    volume_gasoline=gasoline/720000
    g_PM10=volume_gasoline*0.08*1000000
#weight of PM10 calculated for gasoline
    g_PM2=volume_gasoline*0.04*1000000
#weight of PM2.5 calculated for gasoline
    g_NO=volume_gasoline*0.2*1000000
#weight of NO calculated for gasoline
    g_NO2=volume_gasoline*0.1*1000000
#weight of NO2 calculated for gasoline
    g_CO=volume_gasoline*0.8*1000
#weight of CO calculated for gasoline
```

```python
#calculation for coal
def Coal():
    global c_sulfur
    global c_PM10
    global c_PM2
    global c_NO
    global c_NO2
    global c_CO
    global volume_coal
    coal=eval(input("ENTER THE QUANTITY OF COAL USED IN kg"))
    volume_coal=coal/1400000
    c_sulfur=volume_coal*0.08*1000000
#weight of sulfur calculated for coal
    c_PM10=volume_coal*1.5*1000000
#weight of PM10 calculated for coal
    c_PM2=volume_coal*0.8*1000000
#weight of PM2.5 calculated for coal
    c_NO=volume_coal*0.4*1000000
#weight of NO calculated for coal
    c_NO2=volume_coal*0.2*1000000
#weight of NO2 calculated for coal
    c_CO=volume_coal*0.2*1000000
#weight of CO calculated for diesel
```

```python
#calculation for Iron
def Iron():

    iron=eval(input("ENTER THE QUANTITY OF IRON USED IN kg"))
    volume_iron=iron/7870000
    i_sulfur=volume_iron*0.0016*1000000
#weight of sulfur calculated for iron
    i_PM10=volume_iron*0.0063*1000000
#weight of PM10 calculated for iron
    i_PM2=volume_iron*0.0029*1000000
#weight of PM2.5 calculated for iron
    i_NO=volume_iron*0.0012*1000000
#weight of NO calculated for iron
    i_NO2=volume_iron*0.0005*1000000
#weight of NO2 calculated for iron
    i_CO=volume_iron*0.0042*1000
#weight of CO calculated for iron
```

```python
def final():
    Cement()
    Steel()
    Asphalt()
    Diesel()
    LPG()
    Gasoline()
    Coal()
    Iron()
    total_sulfur = ce_sulfur + s_sulfur + a_sulfur + d_sulfur + c_sulfur + i_sulfur
#total weight of sulfur emitted in mu gram
    total_PM10 = ce_PM10 + s_PM10 + a_PM10 + d_PM10 + c_PM10 + i_PM10 + l_PM10 + g_PM10
#total weight of PM10 emitted in mu gram 
    total_PM2 = ce_PM2 + s_PM2 + a_PM2 + d_PM2 + c_PM2 + i_PM2 + l_PM2 + g_PM2
#total weight of PM2 emitted in mu gram 
    total_NO = ce_NO + s_NO + a_NO + d_NO + c_NO + i_NO + l_NO + g_NO
#total weight of NO emitted in mu gram 
    total_NO2 = ce_NO2 + s_NO2 + a_NO2 + d_NO2 + c_NO2 + i_NO2 + l_NO2 + g_NO2
#total weight of NO2 emitted in mu gram 
    total_CO = ce_CO + s_CO + a_CO + d_CO + c_CO + i_CO + l_CO + g_CO
#total weight of CO emitted in mu gram 
##VOLUME 
    volume= (volume_iron + volume_cement + volume_steel + volume_asphalt + volume_diesel + volume_lpg + volume_gasoline + volume_coal)/8
##CONCENTRATION
    S_c=total_sulfur/volume
    PM10_c=total_PM10/volume
    PM2_c=total_PM2/volume
    NO_c=total_NO/volume
    NO2_c=total_NO2/volume
    CO_c=total_CO/volume
    print("Concentration of SO2 is",S_c,'ug/m3')
    print("Concentration of PM10 is",PM10_c,'ug/m3')
    print("Concentration of PM2 is",PM2_c,'ug/m3')
    print("Concentration of NO is",NO_c,'ug/m3')
    print("Concentration of NO2 is",NO2_c,'ug/m3')
    print("Concentration of CO is",CO_c,'mg/m3')
final()
```
